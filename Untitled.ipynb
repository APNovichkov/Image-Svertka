```python
import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test,y_test) = keras.datasets.cifar10.load_data()
```

```python
# Filters

f_outline = [
    [-1,-1,-1],
    [-1,8,-1],
    [-1,-1,-1]
]

f_sharp = [
    [0,-1,0],
    [-1,5,-1],
    [0,-1,0]
]

f_blur = (np.array( [
    [1,1,1],
    [1,1,1],
    [1,1,1]
])/9).tolist()

f_custom = (np.array( [
    [0,  0,  1 ],
    [0, -1,  0],
    [1,  0,  0]
])).tolist()



def apply_filter(img, f):
  ROWS = img.shape[0]
  COLS = img.shape[1]

  d =  np.zeros((ROWS, COLS,3 ), dtype=int).tolist()
  for k in range(3):
    for i in range(ROWS-2):
      for j in range(COLS-2):
        s = 0
        for ii in range(3):
          for jj in range(3):
            s += img[i + ii][j + jj][k]*f[ii][jj]
        d[i+1][j+1][k] = int(s)
  d = np.array(d)
  return np.clip(d, 0, 255)


ROWS = 5
COLS = 9

def add_plot(fig, img, i, j):
  ax = fig.add_subplot(ROWS, COLS, i*COLS + j + 1)
  ax.grid(False)
  ax.axis('off')
  ax.imshow(img)


fig = plt.figure( figsize=(1.5*COLS , 1.5*ROWS))
for i in range(ROWS):
  img_index = np.random.randint(0, x_train.shape[0])  

  j = 0
  img =  x_train[img_index]
  add_plot(fig, img, i,j)

  j = 1
  img = apply_filter(img, f_blur)
  add_plot(fig, img, i,j)

  j = 2
  img = apply_filter(img, f_blur)
  add_plot(fig, img, i,j)

  j = 3
  img = apply_filter(img, f_blur)
  add_plot(fig, img, i,j)

  j = 4
  img =  x_train[img_index]
  img = apply_filter(img, f_sharp)
  add_plot(fig, img, i,j)

  j = 5
  img = apply_filter(img, f_sharp)
  add_plot(fig, img, i,j)

  j = 6
  img = apply_filter(img, f_sharp)
  add_plot(fig, img, i,j)

  j = 7
  img =  x_train[img_index]
  img = apply_filter(img, f_outline)
  add_plot(fig, img, i,j)
  
  j = 8
  img =  x_train[img_index]
  img = apply_filter(img, f_custom)
  add_plot(fig, img, i,j)

plt.show()
```
